*Given an orbiting bodies perihelion and linear velocity at perihelion then any other property of the orbit can be calculated as follows:*

***a**) Keplers law says that the velocity and distance at the aphelion satisfy $l_1v_1 = l_2v_2$ and at the same time the total energy satisfies*
$$
E = \frac{1}{2}mv^{2} - G\frac{mM}{r}
$$
*where m i shte planets mass, $M = 1.9891\times10^{30}\ kg$ is the mass of the sun, and $G = 6.6738 \times 10^{-11}\ m^{3}kg^{-1}s^{-2}*$ is newtons gravitational constant. Given that energy must be conserved show that $v_2$ is the smaller root of the quadratic equation:*
$$
v_2^2 - \frac{2GM}{v_1l_1}v_2 - \left(v_1^2 - \frac{2GM}{l_1}\right) = 0
$$

*Once we have $v_2$ we can calculate $l_2$ using the relation above.*

To identify the roots of a quadratic equation, we rearrange the above like so, swapping $x$ for $v_2$:
$$
x = \frac{\frac{2GM}{v_1l_1} \pm \sqrt{\left(\frac{2GM}{v_1l_1}\right)^2 + 4\left(v_1^2 - \frac{2GM}{l_1}\right)}}{2}
$$

Simplifying this further by canceling out the factors of 2 and reducing the term inside the radical to $\left(v_1 - \frac{GM}{v_1l_1}\right)^2$ and canceling: 

$$
x = {\frac{GM}{v_1l_1} \pm \left(v_1 - \frac{GM}{v_1l_1}\right) }
$$

The solution given by the plus sign is equal to $v_1$ as the two $\frac{GM}{v_1l_1}$ terms cancel.

$$
\frac{GM}{v_1l_1} + \left(v_1 - \frac{GM}{v_1l_1}\right) > {\frac{GM}{v_1l_1} - \left(v_1 - \frac{GM}{v_1l_1}\right) }
$$ 

Therefore the second solution which results in $v_2$ is the smaller of the two roots.

```python
import numpy as np
```

```python
G = 6.6738e-11
M = 1.9891e30

# Create helper functions for all relevant quantities:
def v2(l1, v1):
    return (G*M/(l1*v1)) - (v1 - (G*M/(l1*v1)))

def l2(l1, v1):
    return l1*v1/v2(l1, v1)
    
def semiMajor(l1, v1):

    return 0.5*(l1 + l2(l1, v1))

def semiMinor(l1, v1):
    return np.sqrt(l1*l2(l1, v1))

def period(l1, v1):
    a = semiMajor(l1, v1)
    b = semiMinor(l1, v1)
    return 2*np.pi*a*b/(l1*v1) / (3600*24*365) # divide by the number of seconds in a year

def eccentricity(l1, v1):
    return (l2(l1, v1) - l1) / (l2(l1, v1) + l1)
    
def orbitalFacts():
    l1 = float(input("enter the distance to the sun"))
    v1 = float(input("enter the velocity at perihelion"))
    T = period(l1, v1)
    e = eccentricity(l1, v1)
    print(f"The distance and velocity at aphelion are {l2(l1, v1):0.4} m and {v2(l1, v1):.4} m/s")
    print(f"The period is {T:.4f} years and the eccentricity is {e:.4f}")
    
```

```python
orbitalFacts()
```

### 2.12: Prime Numbers

```python
# only need to check if it is divisible by other primes
# Only need to check factors up too and including sqrt(n)
# If we find ONE factor, then we know it isn't prime and can move on.

primes = [2]
k = 3
while k <= 10000:
    sqrt = np.sqrt(k)
    for i in primes: 
        if k%i == 0: 
            break
        elif i > sqrt:
            primes.append(k)
            break
    k += 1
print(primes)
```

### 2: HR Diagram


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
stars = np.loadtxt("stars.txt")

fig, ax = plt.subplots(1,1, figsize = (10, 10))
plt.tick_params(labelsize = 14)

ax.scatter(stars[:,0], stars[:,1], s = 5, alpha = 0.4)

ax.annotate('Main Sequence', xy=(5000, 5), xytext=(3000, 2.5), fontsize = 12,
            arrowprops=dict(facecolor='black', width = 0.5, shrink=99))

ax.annotate('0.75 $M_\odot$', xy=(4827, 7.75), xytext=(6800, 9), fontsize = 12,
            arrowprops=dict(facecolor='black', width = 0.5, shrink=99))

ax.annotate('1 $M_\odot$', xy=(5780, 5.5), xytext=(7200, 7.25), fontsize = 12,
            arrowprops=dict(facecolor='black', width = 0.5, shrink=99))

ax.annotate('1.5 $M_\odot$', xy=(7444, 2.8), xytext=(9200, 5.25), fontsize = 12,
            arrowprops=dict(facecolor='black', width = 0.5, shrink=99))

ax.invert_yaxis()
ax.invert_xaxis()
ax.set_xlabel('Temperature [K]', fontsize = 16)
ax.set_ylabel('Apparent Magnitude', fontsize = 16)
ax.set_title('Hertzprung-Russel Diagram of Nearby Stars', fontsize = 14)
```
